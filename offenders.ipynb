# Adult and Juvenile Age Category

```python
AJ1 = pd.DataFrame(pd.read_excel("Data/offenders/Offenders_Adult_and_Juvenile_Age_Category_by_Offense_Category_2021.xlsx"))
AJ1
```

# Data Graph 

Age

```python
import matplotlib.pyplot as plt

a1_all_crime = AJ1.loc[AJ1['Offense_Category'] != "Total"]
a1_all_crime = a1_all_crime.sort_values('Total_Offenders', ascending=False) 
a1_all_crime.plot(kind='bar', x='Offense_Category', y='Total_Offenders')
plt.xticks(rotation=90)
```

```python
a1 = AJ1.loc[AJ1['Offense_Category'] != "Total"]
a1 = a1.sort_values('Adult', ascending=False) 

X = a1['Offense_Category'].values
adult = a1['Adult'].values
juvenile = a1['Juvenile'].values

X_axis = np.arange(len(X)) 
plt.bar(X_axis - 0.2, adult, 0.4, label = 'Adult')
plt.bar(X_axis + 0.2, juvenile, 0.4, label = 'Juvenile')
  
plt.xticks(X_axis, X)
plt.xlabel("Offense")
plt.ylabel("Total Offenders")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
age_plot = Age1.loc[:, Age1.columns.drop(['Offense_Category', 'Total_Offenders', 'Unknown_Age', 'Crimes Against'])]

plt.plot(age_plot.columns, age_plot.iloc[0].values, label='Total')
plt.plot(age_plot.columns, age_plot.iloc[1].values, label='Assault Offenses')
plt.plot(age_plot.columns, age_plot.iloc[19].values, label='Drug/Narcotic Offenses')
plt.plot(age_plot.columns, age_plot.iloc[14].values, label='Larceny/Theft Offenses')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Race

```python
r = Race1.loc[Race1['Offense_Category'] != "Total"]
r = Race1.sort_values('White', ascending=False) 

X = r['Offense_Category'].values
white = r['White'].values
black = r['Black_or_African_American'].values
native = r['American_Indian_or_Alaska_Native'].values
asian = r['Asian'].values
island_native = r['Native_Hawaiian_or_Other_Pacific_Islander'].values


X_axis = np.arange(len(X)) 
plt.bar(X_axis - 0.4, white, 0.4, label = 'White')
plt.bar(X_axis - 0.2, black, 0.4, label = 'Black or African American')
plt.bar(X_axis, native, 0.4, label = 'American Indian or Alaskan Native')
plt.bar(X_axis + 0.2, asian, 0.4, label = 'Asian')
plt.bar(X_axis + 0.4, island_native, 0.4, label = 'Native Hawaiin or Other Pacific Islander')
  
plt.xticks(X_axis, X)
plt.xlabel("Offense")
plt.ylabel("Total Offenders")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
s = Sex1.loc[Sex1['Offense_Category'] != "Total"]
s = Sex1.sort_values('Male', ascending=False) 

X = s['Offense_Category'].values
male = s['Male'].values
female = s['Female'].values

X_axis = np.arange(len(X)) 
plt.bar(X_axis - 0.2, male, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.xlabel("Offense")
plt.ylabel("Total Offenders")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
