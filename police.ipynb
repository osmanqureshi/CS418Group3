```python
import pandas as pd
import numpy as np
```

# Offenses in each City

```python
offense = pd.DataFrame(pd.read_excel("Data/Table_8_Offenses_Known_to_Law_Enforcement_by_State_by_City_2021.xlsx"))
offense
```

```python
offense = offense.drop([0, 1, 2, 5878])
offense
```

```python
offense = offense.rename(columns={"Table 8": "State",
                          "Unnamed: 1": "City",
                          "Unnamed: 2": "Population",
                          "Unnamed: 3": "Violent_crime",
                          "Unnamed: 4": "Murder_and_nonnegligent_manslaughter",
                          "Unnamed: 5": "Rape",
                          "Unnamed: 6": "Robbery",
                          "Unnamed: 7": "Aggravated_assault",
                          "Unnamed: 8": "Property_crime",
                          "Unnamed: 9": "Burglary",
                          "Unnamed: 10": "Larceny-theft",
                          "Unnamed: 11": "Motor_vehicle_theft",
                          "Unnamed: 12": "Arson"})
offense
```

```python
offense = offense.reset_index(drop=True)
offense
```

```python
state = ""
for index, row in offense.iterrows():
    if pd.notnull(row['State']):
        state = row['State']
    else:
        row['State'] = state
        
offense
```

```python
offense = offense.astype({'State':'string', 'City':'string', 'Population':'float', 'Violent_crime':'float', 'Murder_and_nonnegligent_manslaughter':'float', 'Rape':'float', 'Robbery':'float', 'Aggravated_assault':'float', 'Property_crime':'float', 'Burglary':'float', 'Larceny-theft':'float', 'Motor_vehicle_theft':'float', 'Arson':'float'})
offense.dtypes
```

# Police in each City

```python
police = pd.DataFrame(pd.read_excel("Data/Table_78_Full_time_Law_Enforcement_Employees_by_State_by_Cities_2021.xlsx"))
police
```

```python
police = police.drop([0, 1, 2, 8656])
police
```

```python
police = police.rename(columns={"Table 78": "State",
                          "Unnamed: 1": "City",
                          "Unnamed: 2": "Population",
                          "Unnamed: 3": "Total_law_enforcement_employees",
                          "Unnamed: 4": "Total_officers",
                          "Unnamed: 5": "Total_civilians"})
police
```

```python
police = police.reset_index(drop=True)
police
```

```python
state = ""
for index, row in police.iterrows():
    if pd.notnull(row['State']):
        state = row['State']
    else:
        row['State'] = state

police
```

```python
police = police.astype({'State':'string', 'City':'string', 'Population':'float', 'Total_law_enforcement_employees':'int', 'Total_officers':'int', 'Total_civilians':'int'})
police.dtypes
```

# Combine Tables

```python
police_offense = pd.merge(offense, police, how='inner', on=['State', 'City'])
police_offense
```

```python
police_offense = police_offense.drop(['Murder_and_nonnegligent_manslaughter', 'Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny-theft', 'Motor_vehicle_theft', 'Arson', 'Population_y', 'Total_officers', 'Total_civilians'], axis=1)
police_offense
```

```python
police_offense['Total_crime'] = police_offense['Violent_crime'] + police_offense['Property_crime']
police_offense
```

```python
police_offense = police_offense.drop(['Violent_crime', 'Property_crime'], axis=1)
police_offense = police_offense.rename(columns={"Population_x": "Population"})
police_offense
```

```python
police_offense['employees_perCapita'] = police_offense["Total_law_enforcement_employees"] / police_offense["Population"]
police_offense['crime_perCapita'] = police_offense["Total_crime"] / police_offense["Population"]
police_offense
```

# Apply Machine Learning Models

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
# https://www.tutorialspoint.com/scikit_learn/scikit_learn_lasso.htm
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.BayesianRidge.html

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import LinearRegression
from numpy import mean
from numpy import std
from numpy import absolute
import seaborn as sns
from matplotlib.colors import SymLogNorm
```

## Linear Regression

```python
y=police_offense["crime_perCapita"]
X=police_offense["employees_perCapita"]
X = X.values.reshape(-1,1)

Xs_train , Xs_test, ys_train, ys_test = train_test_split(X, y, test_size = 0.20)
model = LinearRegression()
model.fit(Xs_train, ys_train)

ys_predict =  model.predict(Xs_test)

print('CRIM Coefficients: \n', model.coef_)
print()

print('Intercept:', model.intercept_)
print()

print('R-squared: %.2f' % r2_score(ys_test, ys_predict))
print()

print('RMSE:', np.sqrt(mean_squared_error(ys_test, ys_predict)))
print('Mean squared error: %.2f'% mean_squared_error(ys_test, ys_predict))

plt.scatter(Xs_train, ys_train,  color='red')
plt.plot(Xs_test, ys_predict, color='blue', linewidth=2)
plt.xlabel("Employees Per Capita")
plt.ylabel("Crime Per Capita")
plt.xticks(())
plt.yticks(())
plt.show()
```

## Bayesian Regression

```python
y=police_offense["crime_perCapita"]
X=police_offense["employees_perCapita"]
X = X.values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

model = BayesianRidge(compute_score=True, n_iter=30)
model.fit(X_train, y_train)
   
y_predict = model.predict(X_test)
   
print('CRIM Coefficients: \n', model.coef_)
print()

print('Intercept:', model.intercept_)
print()

print('R-squared: %.2f' % r2_score(y_test, y_predict))
print()

print('RMSE:', np.sqrt(mean_squared_error(y_test, y_predict)))
print()
```
